```python
class Tax:
    def calculate_tax(self, amount):
        pass

class IndiaTax(Tax):
    def calculate_tax(self, amount):
        return amount * 0.18  # GST 18%

class USATax(Tax):
    def calculate_tax(self, amount):
        return amount * 0.10  # VAT 10%

amount = 1000

india = IndiaTax()
usa = USATax()

print("India Tax:", india.calculate_tax(amount))
print("USA Tax:", usa.calculate_tax(amount))
```
