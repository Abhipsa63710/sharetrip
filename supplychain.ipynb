```python
import os
class Product:
    def __init__(self, product_id, name, quantity, price):
        self.product_id = product_id
        self.name = name
        self.quantity = int(quantity)
        self.price = float(price)

    def __str__(self):
        return (
            f"{self.name} (ID: {self.product_id}) | "
            f"Stock: {self.quantity} | "
            f"Price: ${self.price:.2f}")


class InventoryManager:
    def __init__(self, file_name="inventory.txt"):
        self.file_name = file_name
        self.inventory = {}
        self.load_from_file()

    def load_from_file(self):
        if os.path.exists(self.file_name):
            with open(self.file_name, "r") as file:
                for line in file:
                    parts = line.strip().split(",")
                    if len(parts) == 4:
                        pid, name, qty, price = parts
                        self.inventory[pid] = Product(
                            pid, name, qty, price
                        )

    def save_to_file(self):
        with open(self.file_name, "w") as file:
            for pid, prod in self.inventory.items():
                file.write(
                    f"{pid},{prod.name},"
                    f"{prod.quantity},{prod.price}\n"
                )

    # Inbound Logistics
    def add_or_update_product(
        self, pid, name, qty, price
    ):
        qty = int(qty)
        price = float(price)

        if pid in self.inventory:
            self.inventory[pid].quantity += qty
            self.inventory[pid].price = price

            print(
                f"Stock updated for {name}. "
                f"New Quantity: "
                f"{self.inventory[pid].quantity}"
            )
        else:
            self.inventory[pid] = Product(
                pid, name, qty, price
            )
            print(f"Added new product: {name}")

        self.save_to_file()

    # Outbound Logistics
    def process_sale(self, pid, qty):
        qty = int(qty)

        if pid not in self.inventory:
            print("Product not found.")
            return

        product = self.inventory[pid]

        if product.quantity < qty:
            print(
                f"Insufficient stock! "
                f"Available: {product.quantity}"
            )
            return

        product.quantity -= qty

        bill = qty * product.price

        print("\nSale Successful!")
        print(f"Product: {product.name}")
        print(f"Quantity Sold: {qty}")
        print(f"Total Bill: ${bill:.2f}")

        self.save_to_file()

    def display_inventory(self):
        print("\n========== INVENTORY LIST ==========")

        if not self.inventory:
            print("Inventory is empty.")
            return

        for product in self.inventory.values():
            print(product)

    def check_low_stock(self, threshold=5):
        print("\n========== LOW STOCK ALERT ==========")

        low_stock_found = False

        for product in self.inventory.values():
            if product.quantity <= threshold:
                print(
                    f"{product.name} "
                    f"(ID: {product.product_id}) "
                    f"has only {product.quantity} units left."
                )
                low_stock_found = True

        if not low_stock_found:
            print("All stock levels are healthy.")

    def view_inventory_dashboard(self):
        print("\n===================================")
        print("      INVENTORY DASHBOARD")
        print("===================================")

        total_products = len(self.inventory)

        total_units = sum(
            product.quantity
            for product in self.inventory.values()
        )

        total_value = sum(
            product.quantity * product.price
            for product in self.inventory.values()
        )

        low_stock_count = sum(
            1
            for product in self.inventory.values()
            if product.quantity <= 5
        )

        print(f"Total Products       : {total_products}")
        print(f"Total Stock Units    : {total_units}")
        print(f"Total Inventory Value: ${total_value:.2f}")
        print(f"Low Stock Products   : {low_stock_count}")

        print("\nProduct Summary")
        print("-----------------------------------")

        if not self.inventory:
            print("Inventory is empty.")
        else:
            for product in self.inventory.values():
                print(
                    f"{product.name:<20}"
                    f"Qty: {product.quantity:<5}"
                    f"Price: ${product.price:.2f}"
                )

        print("===================================")


# Main Program
if __name__ == "__main__":

    manager = InventoryManager()

    #adding items (simulating Real-Time Inputs)
    p_id = input("Enter Product ID : ")
    p_name = input("Enter Product Name : ")
    p_qty = input("Enter Product Quantity : ")
    p_price = input("Enter Product Price : ")

    manager.add_or_update_product("101", "Wireless Mouse", 12, 25.00)
    
    manager.add_or_update_product("102", "HDMI Cable", 3, 10.50)
    
    manager.add_or_update_product("103", "Gaming Keyboard", 2, 80.00)
    
    manager.add_or_update_product("104", "USB Hub", 10, 15.99)

    manager.add_or_update_product(p_id, p_name, int(p_qty), float(p_price))

    manager.display_inventory()

    manager.check_low_stock()
```
