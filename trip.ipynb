```python
class Trip:
    def __init__(self, passenger, distance):
        self.passenger = passenger
        self.distance = distance

    def calculate_fare(self):
        fare = self.distance * 10
        return fare

    def show_trip(self):
        print("Passenger:", self.passenger)
        print("Distance:", self.distance, "km")
        print("Fare:", self.calculate_fare())

trip1 = Trip("Rosy", 15)

trip1.show_trip()
```
